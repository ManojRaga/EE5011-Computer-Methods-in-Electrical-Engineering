```python
from matplotlib import colors
import numpy as np
import matplotlib.pyplot as plt

x=np.loadtxt("error_without_non_cancelling_roots.txt",usecols=0)
ya=np.loadtxt("error_without_non_cancelling_roots.txt",usecols=1)
ys=np.loadtxt("error_without_non_cancelling_roots.txt",usecols=2)
x=np.delete(x,0,0)
ya=np.delete(ya,0,0)
ys=np.delete(ys,0,0)

plt.clf()
plt.title('Order of Error vs Alpha / Float / Less Accurate')
plt.loglog(x,ya,'g',label="Non Cancelling Root")
plt.loglog(x,ys,'r',label="Root Near Cancellation")
plt.legend()
plt.xlabel("Alphas")
plt.ylabel("Order of Error")
plt.grid()
```

```python
x1=np.loadtxt("error_with_non_cancelling_roots.txt",usecols=0)
ya1=np.loadtxt("error_with_non_cancelling_roots.txt",usecols=1)
ys1=np.loadtxt("error_with_non_cancelling_roots.txt",usecols=2)
x1=np.delete(x1,0,0)
ya1=np.delete(ya1,0,0)
ys1=np.delete(ys1,0,0)

plt.clf()
plt.title('Order of Error vs Alpha / Float / More Accurate')
m,=plt.loglog(x1,ya1,'g',label="Root Near Cancellation")
n,=plt.loglog(x1,ys1,'r',label="Non Cancelling Root")
plt.legend(handles=[m,n])
plt.xlabel("Alphas")
plt.ylabel("Order of Error")
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.special as sp
```

### Question 1

```python
def s(x,N):
    res=0.0
    for i in range(N+1):
        res=res+(1/(1+i))*sp.jv(i,x)
    return res
```

### Question 2
Forward Recursion

```python
x = 1.5
```

```python
jn = [0 for _ in range(41)]
jn[0], jn[1] = sp.j0(x), sp.j1(x)
for i in range(2, 41):
    jn[i] = (2*i/x)*jn[i-1] - jn[i-2]
```

```python
error = [0, 0]
sum = 0
for n in range(2, 41):
    sum = sum + jn[n]/(1+n)
    error.append(abs(s(x,n)-sum))

xs = [i for i in range(41)]
plt.clf()
plt.semilogy(xs, error, 'c')
plt.xlabel("n Values")
plt.ylabel(" Order of Error")
plt.title("Error for x=1.5")
plt.grid()
```
